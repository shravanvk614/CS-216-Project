```python
import pandas as pd
import seaborn as sns
data = pd.read_fwf('files/results_pct_20201103.txt', error_bad_lines=False)
data = data["County\tElection Date\tPrecinct\tContest Group ID\tContest Type\tContest Name\tChoice\tChoice Party\tVote For\tElection Day\tOne Stop\tAbsentee"]
data = data.str.split(pat="\t")
concatlength = pd.concat([data, data.str.len()], axis=1)
concatlength = concatlength[concatlength.iloc[:, 1] == 15]
data = concatlength.iloc[:, 0]
```

```python
Nov2020 = pd.DataFrame(data.tolist())
Nov2020.columns = ["County", "Election Date", "Precinct", "Contest Group ID", "Contest Type", "Contest Name", "Choice", "Choice Party", "Vote For", "Election Day", "One Stop", "Absentee by Mail", "Provisional", "Total Votes", "Real Precinct"]
```

```python
governor = Nov2020[Nov2020["Contest Name"] == "NC GOVERNOR"]
president = Nov2020[Nov2020["Contest Name"] == "US PRESIDENT"]
ussenate = Nov2020[Nov2020["Contest Name"] == "US SENATE"]

governor = governor[governor["Choice Party"] != ""]
president = president[president["Choice Party"] != ""]
ussenate = ussenate[ussenate["Choice Party"] != ""]

governor['Total Votes'] = governor['Total Votes'].astype(int)
president['Total Votes'] = president['Total Votes'].astype(int)
ussenate['Total Votes'] = ussenate['Total Votes'].astype(int)
```

```python
prespiv = president.pivot_table("Total Votes", index = "County", columns = "Choice", aggfunc = sum)
govpiv = governor.pivot_table("Total Votes", index = "County", columns = "Choice", aggfunc = sum)
prespiv.reset_index(inplace=True)
govpiv.reset_index(inplace=True)

"""trump_vote = []
for x,v in zip(prespiv['Donald J. Trump'],prespiv['Joseph R. Biden']):
    if x>v:
        trump_vote.append('Trump')
    else:
        trump_vote.append('Biden')
prespiv['election_winner'] = trump_vote
prespiv.to_csv('2020pres_county_votes.csv')"""
```

```python

roy_vote = []
for x,v in zip(govpiv['Roy Cooper'],govpiv['Dan Forest']):
    if x>v:
        roy_vote.append('Cooper')
    else:
        roy_vote.append('Forest')
govpiv['election_winner'] = roy_vote
govpiv.to_csv('2020gov_county_votes.csv')
```

```python
treasurer = Nov2020[Nov2020["Contest Name"] == "NC TREASURER"]
treasurer['Total Votes'] = treasurer['Total Votes'].astype(int)
trespiv = treasurer.pivot_table("Total Votes", index = "County", columns = "Choice", aggfunc = 'sum')
ronnie_vote = []
for x,v in zip(trespiv['Dale R. Folwell'],trespiv['Ronnie Chatterji']):
    if x>v:
        ronnie_vote.append('Folwell')
    else:
        ronnie_vote.append('Chatterji')
trespiv['election_winner'] = ronnie_vote
trespiv.to_csv('2020tres_county_votes.csv')
```

```python
right_election = []
for x in Nov2020['Contest Name']:
    if 'NC STATE SENATE DISTRICT' in x:
        right_election.append(True)
    else:
        right_election.append(False)
Nov2020['is_state'] = right_election
state_senate = Nov2020[Nov2020['is_state']]
state_senate['Total Votes'] = state_senate['Total Votes'].astype(int)
state_senatepiv = state_senate.pivot_table("Total Votes", index = "County", columns = "Choice Party", aggfunc = 'sum')
```

```python
who_vote = []
for x,v in zip(state_senatepiv['DEM'],state_senatepiv['TEP']):
    if x>v:
        who_vote.append('DEM')
    else:
        who_vote.append('REP')
state_senatepiv['winner_party'] = who_vote
state_senatepiv.to_csv('2020state_senate_county_votes.csv')
```
